This dataset contains information about 891 people who were on board the ship when departed on April 15th, 1912. As noted in the description on Kaggle's website, some people aboard the ship were more likely to survive the wreck than others. There were not enough lifeboats for everybody so women, children, and the upper-class were prioritized. Using the information about these 891 passengers, the challenge is to build a model to predict which people would survive based on the following fields:

- **Name** (str) - Name of the passenger
- **Pclass** (int) - Ticket class
- **Sex** (str) - Sex of the passenger
- **Age** (float) - Age in years
- **SibSp** (int) - Number of siblings and spouses aboard
- **Parch** (int) - Number of parents and children aboard
- **Ticket** (str) - Ticket number
- **Fare** (float) - Passenger fare
- **Cabin** (str) - Cabin number
- **Embarked** (str) - Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

### Read in Data

_Lets get into the data. Start by importing some libraries we will need and reading in our data._

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

titanic = pd.read_csv('../titanic.csv')
titanic.head()
```

### Clean continuous variables
1. Fill in missing values for `Age`
2. Combine `SibSp` & `Parch`
3. Drop irrelevant/repetitive variables (`SibSp`, `Parch`, `PassengerId`)

_Now just as a refresher, there are 177 missing values for the `Age` feature (RUN CELL BELOW) and we found that they seemed to be missing at random so we will go ahead and fill those missing values. We also decided that `SibSp` and `Parch` both had to do with family size so we will combine those together into one feature. Finally we will drop `SibSp` and `Parch` because the value they're providing will be captured by the family count feature and leaving them all in together could create some multi-collinearity issues. Finally we will drop the `PassengerId` feature because it's an identifier with no implication on whether somebody survived or not._

#### Fill missing for `Age`

```python
titanic.isnull().sum()
```

_Ok, we can replace `Age` with the mean value by using this `.fillna()` method that we saw before. So we will just tell it that we want to fill those nas with the mean age value by calling titanic[`Age`] and calling the mean method. And like we've seen before, we pass in the `inplace` argument to tell it to alter `titanic` as it stands now without require a new dataframe to be called._

_Now if we just print out the top 10 rows in the dataframe, we can see in the age column that 26.699118 is the replacement for a missing value_

```python
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
titanic.head(10)
```

#### Combine `SibSp` & `Parch`

_As a quick refresher, we looked at these categorical plots back in the "Plotting continuous features" lesson and we learned that `SibSp` and `Parch` have the same relationship to the target variable. Which makes sense and since they're both related to family size - we decided to combine them._

```python
for i, col in enumerate(['SibSp', 'Parch']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=titanic, kind='point', aspect=2, )
```

_Combining them really is as simple as just just adding the two columns together and storing it as `Family cnt`._

```python
titanic['Family_cnt'] = titanic['SibSp'] + titanic['Parch']
```

#### Drop unnnecessary variables

_Lastly, we will drop `PassengerId`, `SibSp`, and `Parch` and we indicate to drop them as columns and alter titanic in place. Then we'll print out the top 5 rows to just look at what our dataset looks like. You can see we have `Family cnt` and all of our dropped features are gone. Now we can see that there's still some categorical features that need to be cleaned up._

```python
titanic.drop(['PassengerId', 'SibSp', 'Parch'], axis=1, inplace=True)
```

```python
titanic.head()
```

### Clean categorical variables
2. Create an indicator variable in place of `Cabin`
3. Convert `Sex` to a numeric variable
4. Drop irrelevant/repetitive variables (`Cabin`, `Embarked`, `Name`, `Ticket`)

#### Fill in missing & create indicator for `Cabin`

_Lets start by creating an indicator for `Cabin`. Back in the "Exploring categorical features" lesson we learned that unlike `Age`, `Cabin` **is not** missing at random. In fact, missing values likely meant these people didn't have a cabin and were **much** less likely to survive. One way we explored that is be checking the survival rate for people with a missing value for `Cabin` and those that didn't. We learned that 2/3 of people aboard that HAD cabins survived while < 30% of those aboard that DID NOT have cabins survived._

```python
titanic.groupby(titanic['Cabin'].isnull())['Survived'].mean()
```

_So we're going to create this really simple Cabin indicator using the `where()` method from numpy that acts as an if statement. So the condition we want to use is whether `Cabin` is null, so titanic[`Cabin`].isnull(). If it is, return a 0, if not return a 1. And we're going to store that in `Cabin ind`. And lets print out the top 5 rows to make sure it worked._

_Sure enough wherever `Cabin` is missing, `Cabin ind` is a 0._

```python
titanic['Cabin_ind'] = np.where(titanic['Cabin'].isnull(), 0, 1)
titanic.head()
```

#### Convert `Sex` to numeric

_Next we need to convert the `Sex` feature from `male` vs `female` to numeric. As we talked about before, a model doesn't know what `male` vs `female` really means, it just knows there are two values for `Sex`. What those values are doesn't really matter - so if we convert it to being numeric it just makes it a little easier for most models to handle._

_So we have a dictionary that does the gender to numeric mapping and now we just need to apply that to the `Sex` column using the `.map()` method in pandas. To map it we call titanic[`Sex`], then the `.map()` method and we tell is to use the `gender num` dictionary to control the mapping. So run that and print out the first few rows. You can see that now instead of `male` and `female` it is 0 and 1._

```python
gender_num = {'male': 0, 'female': 1}

titanic['Sex'] = titanic['Sex'].map(gender_num)
titanic.head()
```

#### Drop unnecessary variables

_Back in the "Plotting categorical features section", we discovered that there were different survival rates with the different levels of teh `Embarked` feature. However, this was not a causal relationship - other features like % of women boarding at each port or % of people that have cabins were the real causal factors and since those are already accounted for in this model, we can drop `Embarked`. This also mitigates any potential impact of multi-collinearity._

_We can also drop `Cabin` now that we have `Cabin ind` that captures the same information._

_Lastly, we can drop `Name` and `Ticket` as they are identification variables without any real determining power on whether somebody survived or not._

_So we will run this cell and print out the top 5 rows. Now we can see we have a nice clean dataset without any irrelevant or repetitive variables and without any missing values._

```python
titanic.drop(['Cabin', 'Embarked', 'Name', 'Ticket'], axis=1, inplace=True)
titanic.head()
```

### Write out cleaned data

_Lastly, lets just write out this cleaned dataset so we can use it moving forward_

```python
titanic.to_csv('../titanic_cleaned.csv', index=False)
```
