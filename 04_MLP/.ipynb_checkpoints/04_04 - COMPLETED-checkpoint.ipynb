## Multilayer Perceptron: Fit and evaluate a model

Using the Titanic dataset from [this](https://www.kaggle.com/c/titanic/overview) Kaggle competition.

In this section, we will fit and evaluate a simple Multilayer Perceptron model.

### Read in Data

_Welcome back to the final lesson in the Multilayer Perceptron chapter. As we saw previously, we're going to use `GridSearchCV`  to do Grid Search within k-fold Cross Validation in order to find the optimal hyperparameter settings for Multilayer Perceptron that generates the best model._

_Lets start by importing the packages we will need for this lesson, then we will read in the training features and the training labels._

```python
import joblib
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier

tr_features = pd.read_csv('../train_features.csv')
tr_labels = pd.read_csv('../train_labels.csv', header=None)
```

### Hyperparameter tuning

![hidden layer](img/hidden_layers.png)
![activation](img/activation.png)
![learning rate](img/lr.png)

_This is just a quick reminder of the three hyperparameters we will seek to optimize in this section:_
1. _What is the right size of the hidden layer or layers?_
2. _What is the right activation function for this problem?_
3. _What is the optimal learning rate to find the optimal model for this problem?_

_Here is that `print results` function that is just going to help us explore the results of this GridSearchCV._

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

_So we have called the `MLPClassifier` object and stored it as `mlp`. We could hardcode hyperparamter values in here by entering them in the parentheses but we want to test that using `GridSearchCV` and we'll just leave the rest of the hyperparameter settings as the default setting we saw in the last lesson._

_Lets define our hyperparameter dectionary. So the keys in this dictionary align with the name of the hyperparameters that would be passed into `MLPClassifier` and then we just need to define the list of settings we want to test._

_So for hidden layer size, we'll only be exploring models with ONE hidden layer. So we'll look at one layer with 10 nodes, one layer with 50 nodes, and one layer with 100 nodes. The second entry in this tuple would represent the number of layers._

_Next, we will explore the three activation functions we looked at in the last lesson - so that's `relu` (this is the default), `tanh`, and `logistic` which is the same as the sigmoid we saw before._

_Lastly, we'll look at the learning rate. Now here are a few options given in `sklearn` and try to view this using the image above to understand what this actually means:_
- _`constant` is just keeping a constant `learning rate` given by a different hyperparameter_
- _`invscaling` gradually decreases the learning rate at each step, so this will allow it to take large jumps at first and then slowly decrease it hopefully as it gets closer and closer to the optimal model_
- _`adaptive` keeps the learning rate constant as long as training loss keeps decreasing, if it STOPS decreasing then it will decrease the learning rate to take smaller steps_

_So now call `GridSearchCV`, pass in our model object (`mlp`), the hyperparameter dictionary, and tell it we want to do 5-fold Cross Validation._

_We just call `.fit()` with our training features and labels and that will prompt it to grab the first hyperparameter combination, so that would be one hidden layer with 10 nodes, using the `relu` activation function, and a constant learning rate. And it will use that setting and run Cross Validation. It will do this for each hyperparameter combination._

_So lets look at the results._

```python
mlp = MLPClassifier()
parameters = {
    'hidden_layer_sizes': [(10,), (50,), (100,)],
    'activation': ['relu', 'tanh', 'logistic'],
    'learning_rate': ['constant', 'invscaling', 'adaptive']
}
cv = GridSearchCV(mlp, parameters, cv=5)
cv.fit(tr_features, tr_labels)

print_results(cv)
```

_Lets take a quick look at the best estimator based on test score_

```python
cv.best_estimator_
```

### Write out pickled model

_Lastly, lets write out this model to compare it to the other models a little later on. One last reminder, cv.best estimator is a MODEL - there is an actual equation stored. So we can save it, then read it back in and go right to work making predictions with that model._

_We're going to pickle the model and write it out using `joblib`. Pass in the model object and just tell it to write to the same location where we stored our other model `../MLP model.pkl`._

```python
joblib.dump(cv.best_estimator_, '../MLP_model.pkl')
```
