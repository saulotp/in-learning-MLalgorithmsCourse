## Boosting: Fit and evaluate a model

Using the Titanic dataset from [this](https://www.kaggle.com/c/titanic/overview) Kaggle competition.

In this section, we will fit and evaluate a simple Gradient Boosting model.

### Read in Data

_Welcome back to the FINAL lesson before we compare all of these algorithms to one another. In this lesson we're going to use `GridSearchCV`  to do Grid Search within k-fold Cross Validation in order to find the optimal hyperparameter settings for Gradient Boosting that generates the best model._

_Lets import the packages we will need for this lesson, then we will read in the training features and the training labels._

```python
import joblib
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

tr_features = pd.read_csv('../train_features.csv')
tr_labels = pd.read_csv('../train_labels.csv', header=None)
```

### Hyperparameter tuning

![GB](img/gb.png)
![LR](img/lr.png)

_A quick reminder of what the three hyperparameters we will be tuning represent. Number of estimators simply represents how many individual decision trees to build, max depth dictates how deep each of those trees can go, and learning rate controls how quickly this algorithm will try to find the optimal model._

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

_So we have called the `GradientBoostingClassifier` object and stored it as `gb`. Leave the parentheses blank as we have in each chapter._

_Lets define our hyperparameter dectionary. So the keys in this dictionary align with the name of the hyperparameters that would be passed into `GradientBoostingClassifier` and then we just need to define the list of settings we want to test._

_So for number of estimators we will test out building 5, 50, 250, and 500 decision trees._

_And then for max depth we will start with 1 (what's called a decision stump) and then we'll also test out 3, 5, 6, and 9. One thing I want to call your attention to is that we're testing MORE individual trees but more shallow trees than we did with Random Forest. Again, this goes back to how each of the two algorithms optimize the bias/variance trade-off to find the best model. Gradient Boosting does better with a lot of shallow trees while Random Forest goes deeper on fewer trees._

_So now call `GridSearchCV`, pass in our model object (`gb`), the hyperparameter dictionary, and tell it we want to do 5-fold Cross Validation._

_We just call `.fit()` with our training features and labels like we have in every other chapter and that will run 5-fold Cross Validation for each hyperparameter setting combination._

_So lets look at the results._

```python
gb = GradientBoostingClassifier()
parameters = {
    'n_estimators': [5, 50, 250, 500],
    'max_depth': [1, 3, 5, 7, 9],
    'learning_rate': [0.01, 0.1, 1, 10, 100]
}
cv = GridSearchCV(gb, parameters, cv=5)
cv.fit(tr_features, tr_labels)

print_results(cv)
```

_Lets take a quick look at the best estimator based on test score_

```python
cv.best_estimator_
```

### Write out pickled model

_Lastly, lets write out this model to compare it to the other models in the next chapter. We just call `joblib.dump()` and pass in the model object and just tell it to write to the same location where we stored our other models `../GB model.pkl`._

```python
joblib.dump(cv.best_estimator_, '../GB_model.pkl')
```
